```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
```

# Set up the Diffusion Pipeline

```python
pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()
```

# Define a prompt for small video generation

```python
prompt_small = "Orangutan is playing"
```

# Generate small video

```python
video_frames_small = pipe(prompt_small, num_inference_steps=50).frames
video_path_small = export_to_video(video_frames_small)
video_name_small = video_path_small.replace('/tmp', '')
```

### Display small video details


```python
print('Small Video Name:', video_name_small)
torch.cuda.empty_cache()
```

# Define a prompt for large video generation


```python
prompt_large = "Penguins plan a surprise party for a polar bear"
```

```python
pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
```

### Optimize for GPU memory


```python
pipe.enable_model_cpu_offload()
pipe.enable_vae_slicing()
```

# Generate large video

```python
video_frames_large = pipe(prompt_large, num_inference_steps=25, num_frames=100).frames
video_path_large = export_to_video(video_frames_large)
video_name_large = video_path_large.replace('/tmp', '')
```

### Display large video details


```python
print('Large Video Name:', video_name_large)
torch.cuda.empty_cache()
```
